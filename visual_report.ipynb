### Most Common Dog Stage

One of the more interesting and unique aspects of WeRateDogs is the lingo that has developed around the account. Dogs are called 'pupper', 'puppo', or 'doggo' based on how old their dog is with 'pupper' being used for puppies, 'doggo' for adult dogs, and 'puppo' for dogs in a stage of development between puppy and adult.
```python
df['stage'].value_counts().head(5).plot(kind='bar')
plt.title('Most Common Dog Stage');
plt.xlabel('Dog Stage');
plt.ylabel('Count');
```
